```python
import pandas as pd

# Baca data transaksi
data = pd.read_excel("Homework-DataFrame-Dataset (1).xlsx", sheet_name="Paid-Transaction")

# Gabungkan first name dan last name, ubah ke Title Case
data["Full Name"] = (data["First Name"] + " " + data["Last Name"]).str.title()

# Atur posisi kolom agar Full Name ada setelah Last Name
kolom = list(data.columns)
pos = kolom.index("Last Name") + 1
kolom.insert(pos, kolom.pop(kolom.index("Full Name")))
data = data[kolom]

# Cek hasilnya
data[["First Name", "Last Name", "Full Name"]].head()
```

```python
# Hitung nilai GMV = Transaction Amount - Seller Discount + Delivery Fee
data["GMV"] = data["Transaction Amount"] - data["Seller Discount"] + data["Delivery Fee"]

# Pindahkan kolom GMV agar berada setelah Seller Discount
kolom2 = list(data.columns)
letak = kolom2.index("Seller Discount") + 1
kolom2.insert(letak, kolom2.pop(kolom2.index("GMV")))
data = data[kolom2]

# Tampilkan contoh data
data[["Transaction Amount", "Seller Discount", "Delivery Fee", "GMV"]].head()
```

```python
# Buat dictionary untuk mapping Meta Category ke Group
kategori = {
    "Home/ Food": "Group 1",
    "Sports/ Hobi/ Otomotif": "Group 1",
    "Fashion": "Group 1",
    "Beauty/ Health": "Group 1",
    "Gadget/ Komputer": "Group 2",
    "Elektronik": "Group 2",
    "Service/ Mokado": "Group 2",
    "Babies/ Kids": "Group 2"
}

# Tambahkan kolom Group berdasarkan mapping
data["Group"] = data["Meta Category"].map(kategori)

# Tambahkan kolom bulan dari tanggal transaksi
data["Month"] = pd.to_datetime(data["Paid Date"], format='%Y%m%d').dt.to_period("M")

# Buat pivot table: GMV total tiap bulan per Group
pivot = data.pivot_table(index="Month", columns="Group", values="GMV", aggfunc="sum").fillna(0)

# Tampilkan hasilnya
display(pivot)
```

```python
# Baca sheet Seller
seller = pd.read_excel("Homework-DataFrame-Dataset (1).xlsx", sheet_name="Seller")

# Gabungkan data transaksi dengan seller
gabung = pd.merge(data, seller, on="Order Number", how="left")

# Ubah tanggal ke period bulan
gabung["Month"] = pd.to_datetime(gabung["Paid Date"], format='%Y%m%d').dt.to_period("M")

# Ambil data Agustus 2017
agustus = gabung[gabung["Month"] == "2017-08"]

# Hitung total GMV per seller
total_gmv = agustus.groupby("Seller")["GMV"].sum().sort_values(ascending=False)

# Tampilkan seller dengan GMV tertinggi
total_gmv.head(1)
```

```python
# Filter data bulan September dan Meta Category Fashion
sept_fashion = gabung[
    (gabung["Month"] == "2017-09") &
    (gabung["Meta Category"] == "Fashion")
]

# Hitung jumlah transaksi (count) per seller
jumlah = sept_fashion.groupby("Seller")["Order Number"].count().sort_values(ascending=False)

# Tampilkan seller dengan transaksi terbanyak
jumlah.head(1)
```
